## Data Cleansing

```python
data_dir = './fixed_sample/old_training/'
fixed_source = torch.load(os.path.join(data_dir, 'fixed_source_6.pkl'))
```

```python
plt.figure(figsize=(14, 18))
for idx, image in enumerate(fixed_source):
    plt.subplot(8, 6, idx+1)
    plt.imshow(image.cpu().detach().numpy().reshape(128, 128), cmap='gray')
#     plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(14, 18))
for idx, image in enumerate(fixed_source):
    image = centering_image(image.cpu().detach().numpy().reshape(128, 128), resize_fix=90)
    plt.subplot(8, 6, idx+1)
    plt.imshow(image, cmap='gray')
#     plt.axis('off')
plt.show()
```
